# Merging DataFrames

```python
df4 = pd.merge(df1, df2, on = 'tconst')
df4
```

```python
df = pd.merge(df4, df3, left_on = 'primary_title', right_on = 'title')
df
```

# Dropping NaN

```python
df.dropna(subset = ['genres', 'averagerating'], inplace=True)
df
```

# Getting Unique Genres

```python
genre = str(df['genres'].unique())
```

```python
genre = set(genre.replace("[","").replace("]","").replace("\n", "").replace("'", "").replace(" ", ",").split(","))
```

# Creating Genre Rating Dictionary 

genre_rating_dict = {}
for g in genre:

```python
fig, ax = plt.subplots(figsize = (11,11))
x = genre_rating_dict.keys()
y = genre_rating_dict.values()
color = ["red", "orangered", "maroon", "yellow", "gold", "orange", "crimson", "chartreuse", "green", "lime", 
         "cyan", "blue", "deepskyblue", "royalblue", "fuchsia", "pink", "deeppink", "indigo", 
         "lavender", "darkviolet", "silver", "black"]
plt.bar(x, y, color = color)

ax.set_title('Genres and Ratings', fontsize = 27);
ax.set_xlabel('Genres', fontsize = 15)
ax.set_ylabel('Ratings', fontsize = 15)
ax.set_ylim(0, 10)
ax.set_xticklabels(genre_rating_dict.keys(), rotation = 'vertical')

plt.savefig('genres')
```

```python
genre_rating_df = pd.DataFrame(genre_rating_dict.items(), columns= ['genres', 'averagerating']) 
genre_rating_df
```

Changing df to q2 so I can drop different rows and not drop NaN in ratings

```python
q2 = pd.merge(df4, df3, left_on = 'primary_title', right_on = 'title')
q2
```

```python
q2.dropna(subset = ['genres', 'domestic_gross', 'foreign_gross'], inplace=True)
q2
```

```python
q2.dtypes
```

```python
q2['foreign_gross'] = q2['foreign_gross'].apply(lambda x: float(x.replace(',','')))
```

```python
q2['total_gross'] = q2['domestic_gross'] + q2['foreign_gross']
q2['total_gross']
```

```python
q2.info()
```

```python
genre_gross_dict = {}
for g in genre:
    genre_gross_dict[g] = q2[q2['genres'].str.contains(g)]['total_gross'].mean()
print(genre_gross_dict)
```

```python
fig, ax = plt.subplots(figsize = (11,11))
x = genre_gross_dict.keys()
y = genre_gross_dict.values()
color = ["red", "orangered", "maroon", "yellow", "gold", "orange", "crimson", "tan", "chartreuse", "green", "lime", 
         "turquoise", "cyan", "blue", "deepskyblue", "royalblue", "fuchsia", "pink", "deeppink", "indigo", "purple", 
         "lavender", "darkviolet", "grey", "silver", "black"] 
plt.bar(x, y, color = color)

ax.set_title('Genres and Revenue');
ax.set_xlabel('Genres')
ax.set_ylabel('Revenue (dollars)')
ax.set_ylim(0, 400000000)
ax.set_yticklabels(['0', '$ 50,000,000', '$ 100,000,000', '$ 150,000,000', '$ 200,000,000','$ 250,000,000',
                    '$ 300,000,000', '$ 350,000,000', '$400,000,000'])

ax.set_xticklabels(genre_gross_dict.keys(), rotation = 'vertical')

plt.savefig('genre_money')
```
